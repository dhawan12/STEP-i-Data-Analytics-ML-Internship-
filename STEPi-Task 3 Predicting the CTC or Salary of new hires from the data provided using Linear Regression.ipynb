```python
del test['Role']
del test['College']
del test['City type']
```

```python
test.head()
```

```python
x=test[['previous CTC','previous job changes','Graduation marks','College_T1','College_T2','Role_Manager','City_Metro']]
y=test[['Actual CTC']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predict=lr.predict(x)
test['Predicted CTC']=predict
test.head()
```

```python
predict
```

```python
sns.lmplot(x='Actual CTC', y='Predicted CTC', data=test, line_kws={'color':'pink'})
```

```python
from sklearn import metrics
print("Mean Squared Error: ",metrics.mean_squared_error(y,predict))
print("Root Mean Squared Error: ",np.sqrt(metrics.mean_squared_error(y,predict)))
```
